```python

import re

# Function to extract discount percentage or flat amount and categorize it
def extract_discount(offer):
    # Check for percentage-based discounts
    percentage_match = re.search(r'(\d+)%', offer)
    
    if percentage_match:
        discount = int(percentage_match.group(1))  # Extract percentage discount
        discount_type = 'Percentage'
    else:
        # Check for flat amount discount like "50 off Promo"
        flat_match = re.search(r'(\d+)\s*off', offer)
        discount = int(flat_match.group(1)) if flat_match else 0  # Extract flat discount
        discount_type = 'Flat'

    # Categorize the type of offer
    if 'New User' in offer:
        category = 'New User'
    elif 'App' in offer:
        category = 'App'
    elif 'Promo' in offer:
        category = 'Promo'
    else:
        category = 'Normal Discount'  # For cases like '10% normal discount'
    
    return pd.Series([discount, discount_type, category])

# Apply the function to extract discount details
df[['Discount Value', 'Discount Type', 'Discount Category']] = df['Discounts and Offers'].apply(extract_discount)

# Function to calculate the final order value after applying the discount
def calculate_discounted_value(row):
    if row['Discount Type'] == 'Percentage':
        return row['Order Value'] * (1 - row['Discount Value'] / 100)
    else:  # Flat discount
        return row['Order Value'] - row['Discount Value']

# Apply the calculation to create the 'Discounted Order Value' column
df['Discounted Order Value'] = df.apply(calculate_discounted_value, axis=1)

# Show the resulting DataFrame with discount columns
df.head()



```

```python
# Calculate profit for each order
df['Profit'] = df['Discounted Order Value'] - (df['Commission Fee'] + df['Payment Processing Fee'])

# Total profit
total_profit = df['Profit'].sum()

df.head()
```

```python
# Calculate total profit per customer
customer_profit = df.groupby('Customer ID')['Profit'].sum().sort_values(ascending=False)
customer_profit.head()

```

```python
# Average order value
avg_order_value = df['Order Value'].mean()
print(f"Average Order Value: {avg_order_value}")

# Average profit per order
avg_profit_per_order = df['Profit'].mean()
print(f"Average Profit per Order: {avg_profit_per_order}")
```

```python
# Group by discount type and calculate average profit
discount_profit_impact = df.groupby('Discount Type')['Profit'].mean()
print(discount_profit_impact)
```

```python
# Calculate total refunds
total_refunds = df['Refunds/Chargebacks'].sum()
print(f"Total Refunds: {total_refunds}")

# Calculate profit after refunds
df['Profit After Refunds'] = df['Profit'] - (df['Refunds/Chargebacks'] * avg_order_value)
total_profit_after_refunds = df['Profit After Refunds'].sum()
print(f"Total Profit After Refunds: {total_profit_after_refunds}")
```

```python

df['Order Date and Time'] = pd.to_datetime(df['Order Date and Time'], format='%Y-%m-%d %H:%M:%S')
df['Delivery Date and Time'] = pd.to_datetime(df['Delivery Date and Time'], format='%Y-%m-%d %H:%M:%S')

# Step 2: Calculate the total delivery time in minutes
df['Delivery Time Taken (Minutes)'] = (df['Delivery Date and Time'] - df['Order Date and Time']).dt.total_seconds() / 60

# Print the DataFrame with the delivery time in minutes
df[['Order Date and Time', 'Delivery Date and Time', 'Delivery Time Taken (Minutes)']].head()
```

```python
df['Profit Margin (%)'] = (df['Profit'] / df['Order Value']) * 100
df['Profit Margin (%)'] = df['Profit Margin (%)'].round(2)
```

```python
# Ensure 'Order Date and Time' column is in datetime format
df['Order Date and Time'] = pd.to_datetime(df['Order Date and Time'], errors='coerce')

# Extract week number from 'Order Date and Time' and add 'Week' as prefix
df['Order Week'] = 'Week ' + df['Order Date and Time'].dt.isocalendar().week.astype(str)

# Show the resulting DataFrame with the 'Order Week' column
df[['Order Week']].head()
```
